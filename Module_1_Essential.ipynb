```python
az_zipcodes = pd.read_csv("data.csv")
az_zipcodes.head()
```

```python
zip_code = az_zipcodes['zip'].tolist()
```

```python
# Openweather API key
api_key = '545243cdc0f0dfaff9872d7bccff0839'

#geo_base_url variable to store url
geo_base_url = 'http://api.openweathermap.org/geo/1.0/zip?'

# base_url variable to store url
base_url = 'https://api.openweathermap.org/data/2.5/weather?'
```

```python
with open('az_zip_geo.csv','w') as az_zip_geo, open('az_weather.csv','w') as az_weather:
    for zipcode in zip_code:
        
        sleep(10)
        
        geo_complete_url = geo_base_url + 'zip=' + str(zipcode) + '&appid=' + api_key
        results = requests.get(geo_complete_url)
        data = results.json()
        
        lineout = data['name'] + ',' + str(data['zip']) + ',' + str(data['lat']) + ',' + str(data['lon']) + '\n'
        
        lat = str(data['lat'])
        lon = str(data['lon'])
        
        
        az_zip_geo.write(lineout)
        
        
            
        sleep(10)
            
        complete_url = base_url + "lat=" + lat + "&lon=" + lon + "&units=imperial" + "&appid=" + api_key
        results = requests.get(complete_url)
        data = results.json()
                
        lineout = data['name'] + ','
        lineout += str(data['coord']['lat']) + ',' + str(data['coord']['lon']) + ','
        lineout += data['weather'][0]['description'] + ','
        lineout += str(data['main']['temp']) + ','
        lineout += str(data['main']['humidity']) + ','
        lineout += datetime.strftime((datetime.fromtimestamp(data['dt'])),'%Y-%m-%d %I:%M %p') + '\n'
                
        print(lineout)
        az_weather.write(lineout)
```
